## Notes:
### Categorical: 
- season: 1: winter, 2: spring, 3: summer, 4: fall ; OneHot
- yr: 0: 2011, 1: 2012
- workingday: if day is neither weekend nor holiday is 1, otherwise is 0
- holiday: weather day is holiday:1 or not: 0
- weathersit: 1: Clear / Few clouds, 2: Mist + Cloudy, 3: Light Snow / Light Rain, 4: Heavy Rain / Ice / Thunderstorm / Snow + Fog; OneHot
### Cyclic:
- mnth: month (1 to 12)
- hr: hour(0 to 23)
- weekday: day of the week(0 to 6)

## EDA

```python
sns.histplot(df['cnt'], bins=30, kde=True, color='#320f3d')
plt.title("Distribution of Hourly Bike Rentals")
plt.xlabel("Bike Count (cnt)")
plt.show()
```

```python
# Create a custom cubehelix palette
custom_cube = sns.cubehelix_palette(8)[4:6]
ax = sns.lineplot(
    x='mnth', y='cnt', hue='yr', data=df,
    estimator='mean',
    palette=custom_cube
)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['2011', '2012'], title='Year')

plt.title("Average Hourly Bike Rentals by Month")
plt.xlabel("Month")
plt.ylabel("Average Count")
plt.tight_layout()
plt.show()
```

```python
custom_cube = sns.cubehelix_palette(12)
ax = sns.lineplot(x='hr', y='cnt', hue='mnth', data=df, estimator='mean', palette=custom_cube)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, 
          ["Jan","Feb","Mar","Apr","May","Jun",
           "Jul","Aug","Sep","Oct","Nov","Dec"],
          title="Month")

plt.title("Hourly Bike Rentals Across Months")
plt.xlabel("Hour")
plt.ylabel("Average Count")
plt.tight_layout()
plt.show()
```

```python
custom_cube = sns.cubehelix_palette(6)
sns.boxplot(x='workingday', y='cnt', data=df, color='#613969')
plt.title("Bike Rentals: Working Day vs Non-Working Day")
plt.xlabel("Working Day (1 = Yes)")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='season', y='cnt', data=df, color='#613969')
plt.title("Bike Rentals Across Seasons")
plt.xlabel("Season")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='weathersit', y='cnt', data=df, color='#613969')
plt.title("Bike Rentals vs Weather Situation")
plt.xlabel("Weather Situation")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='temp', y='cnt', data=df, alpha=0.3, color='#613969')
plt.title("Bike Rentals vs Temperature")
plt.xlabel('Temperature')
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='hum', y='cnt', data=df, alpha=0.3, color='#613969')
plt.title("Bike Rentals vs Humidity")
plt.xlabel('Humidity')
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='windspeed', y='cnt', data=df, alpha=0.3, color='#613969')
plt.title("Bike Rentals vs Humidity")
plt.xlabel('Wind Speed')
plt.tight_layout()
plt.show()
```

```python
num_cols = ['cnt', 'temp', 'atemp', 'hum', 'windspeed']
custom_cube = sns.cubehelix_palette(5)
sns.heatmap(df[num_cols].corr(), annot=True, cmap=custom_cube)
plt.title("Correlation Matrix (Numeric Features)")
plt.tight_layout()
plt.show()
```
