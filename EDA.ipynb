## Notes:
### Categorical: 
- season: 1: winter, 2: spring, 3: summer, 4: fall ; OneHot
- yr: 0: 2011, 1: 2012
- workingday: if day is neither weekend nor holiday is 1, otherwise is 0
- holiday: weather day is holiday:1 or not: 0
- weathersit: 1: Clear / Few clouds, 2: Mist + Cloudy, 3: Light Snow / Light Rain, 4: Heavy Rain / Ice / Thunderstorm / Snow + Fog; OneHot
### Cyclic:
- mnth: month (1 to 12)
- hr: hour(0 to 23)
- weekday: day of the week(0 to 6)

## EDA

```python
sns.histplot(df["cnt"], bins=30, kde=True, color="#320f3d")
plt.title("Distribution of Hourly Bike Rentals")
plt.xlabel("Bike Count (cnt)")
plt.ylabel("Frequency")
plt.tight_layout()

# ---------- Save: transparent background ----------
plt.savefig(
    "assets/bike_rentals_distribution_transparent.svg",
    format="svg",
    transparent=True
)

# ---------- Save: solid background ----------
plt.savefig(
    "assets/bike_rentals_distribution_bg.svg",
    format="svg",
    facecolor="white"
)

plt.show()
```

```python
# Create a custom cubehelix palette
custom_cube = sns.cubehelix_palette(8)[4:6]

ax = sns.lineplot(
    x="mnth",
    y="cnt",
    hue="yr",
    data=df,
    estimator="mean",
    palette=custom_cube
)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ["2011", "2012"], title="Year")

plt.title("Average Hourly Bike Rentals by Month")
plt.xlabel("Month")
plt.ylabel("Average Count")
plt.tight_layout()

# ---- Save SVGs ----
plt.savefig(
    "assets/avg_hourly_bike_rentals_by_month_transparent.svg",
    format="svg",
    transparent=True
)

plt.savefig(
    "assets/avg_hourly_bike_rentals_by_month_bg.svg",
    format="svg",
    facecolor="white"
)

plt.show()
```

```python
custom_cube = sns.cubehelix_palette(12)

ax = sns.lineplot(
    x="hr",
    y="cnt",
    hue="mnth",
    data=df,
    estimator="mean",
    palette=custom_cube
)

handles, labels = ax.get_legend_handles_labels()
ax.legend(
    handles,
    ["Jan","Feb","Mar","Apr","May","Jun",
     "Jul","Aug","Sep","Oct","Nov","Dec"],
    title="Month"
)

plt.title("Hourly Bike Rentals Across Months")
plt.xlabel("Hour")
plt.ylabel("Average Count")
plt.tight_layout()

# ---- Save SVGs ----
plt.savefig(
    "assets/hourly_bike_rentals_across_months_transparent.svg",
    format="svg",
    transparent=True
)

plt.savefig(
    "assets/hourly_bike_rentals_across_months_bg.svg",
    format="svg",
    facecolor="white"
)

plt.show()
```

```python
num_cols = ["cnt", "temp", "atemp", "hum", "windspeed"]
custom_cube = sns.cubehelix_palette(5)

corr = df[num_cols].corr()

# Create mask for upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    cmap=custom_cube,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)

plt.title("Correlation Matrix (Numeric Features)")
plt.tight_layout()

# ---- Save SVGs ----
plt.savefig(
    "assets/correlation_matrix_numeric_lower_triangular_transparent.svg",
    format="svg",
    transparent=True
)

plt.savefig(
    "assets/correlation_matrix_numeric_lower_triangular_bg.svg",
    format="svg",
    facecolor="white"
)

plt.show()
```
